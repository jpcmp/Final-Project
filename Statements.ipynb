```python
import os
import shutil
import datetime
import pandas as pd
import numpy as np
import win32com.client as win32
from PyPDF2 import PdfFileReader, PdfFileWriter

blackbook = pd.read_csv('I:/belz/data/Blackbook_v1.1/blackbook_' + datetime.datetime.today().strftime('%Y%m%d') + '.csv')
blackbook['BOI_CODE']=blackbook['BOI_CODE'].astype(str)
statementpw = pd.read_excel('//NBNOTOOPS1/opspublic/Trade Support Desk (Cash)/Client Statements/Client Statement PW file.xlsx')
merge = pd.merge(left=blackbook[['COUNTERPARTY_CODE', 'BOI_CODE']], right=statementpw, how='left', left_on='COUNTERPARTY_CODE', right_on='FDA NAME')
group = pd.DataFrame() 

path = '//NBNOTOOPS1/opspublic/Trade Support Desk (Cash)/Client Statements/'
files = [f for f in os.listdir(path+'New Statements/') if '.pdf' in f]
for file in iter(files):
    shutil.move(os.path.join(path+'New Statements/',file),os.path.join(path+'New Statements/','e'+file))
    with open(path+'New Statements/e'+file, "rb") as in_file:
        input_pdf = PdfFileReader(in_file)
        output_pdf = PdfFileWriter()
        output_pdf.appendPagesFromReader(input_pdf)
        acct = ''
        for c in str(file):
            if c.isdigit(): acct = acct+str(c)
        acctref = merge.loc[merge['BOI_CODE'].str.startswith(acct[:8])].reset_index()
        try:
            password = acctref['PASSWORD'][0]
        except IndexError:
            print('Acct ' + acct[:8] + ' could not be found in the Blackbook')
            next(iter(files), None)
            continue
        try:
            output_pdf.encrypt(password)
        except TypeError:
            print('Acct ' + acct[:8] + ' is under ' + acctref['COUNTERPARTY_CODE'][0] + ' who is a new client to our List')
            next(iter(files), None)
            continue
        with open(path+'New Statements/'+file, "wb") as out_file:
            output_pdf.write(out_file)
            acctref['out_file'] = file
        group = group.append(acctref.loc[0])
    
    os.remove(os.path.join(path+'New Statements/','e'+file))
    password = None
    
clients = group['CLIENT NAME'].unique()
for client in clients:
    filtered = group.loc[group['CLIENT NAME'] == client].reset_index()
    outlook = win32.Dispatch('outlook.application')
    mail = outlook.CreateItem(0)
    mail.SentOnBehalfOfName = 'Account, NB IE Toronto Support'
    mail.Display()
    signature = mail.HTMLBody
    if pd.notnull(filtered['EMAILS'][0]): mail.To = filtered['EMAILS'][0]
    mail.Subject = 'BMO Statement'
    mail.HTMLBody = '<h2><font size="4" style="font-weight:normal">Hi,<br><br> Please see your statements attached.<br><br>Thanks,</font></h2><h3>' + signature + '</h3>'
    for att in filtered['out_file']:
        attachment = path+'New Statements/'+att
        mail.Attachments.Add(attachment)
    mail.Display()
    outlook = None
    outlook = win32.Dispatch('outlook.application')
    mail = outlook.CreateItem(0)
    mail.SentOnBehalfOfName = 'Account, NB IE Toronto Support'
    mail.Display()
    signature = mail.HTMLBody
    if pd.notnull(filtered['EMAILS'][0]): mail.To = filtered['EMAILS'][0]
    mail.Subject = 'BMO Statement'
    mail.HTMLBody = '<h2><font size="4" style="font-weight:normal">Hi,<br><br> You will be receiving your statements in a few minutes. Please find your password to have access to the files below:<br><br>' + filtered['PASSWORD'][0] + '<br><br>Thanks,<br></font></h2><h3>' + signature + '</h3>'
    mail.Display()
    outlook = None

    
# Moving .msg files to Old Confirms folder 
for file in files:
    if os.path.isfile(os.path.join(path, file)):
        os.remove(os.path.join(path, file))
    try:
        shutil.move(os.path.join(path+'New Statements/', file), os.path.join(path, file))        
    except:
        shutil.move(os.path.join(path+'New Statements/','e'+file),os.path.join(path+'New Statements/',file))
```
